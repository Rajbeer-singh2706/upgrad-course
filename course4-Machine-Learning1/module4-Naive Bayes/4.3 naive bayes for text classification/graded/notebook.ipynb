```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Load the dataset
train_data = pd.read_csv('./movie_review_train.csv')
train_data.head()
```

```python
test_data = pd.read_csv('movie_review_test.csv')
test_data.head()
```

```python
# Preprocess the data (basic cleaning)
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation and special characters
    text = ''.join([char for char in text if char.isalnum() or char.isspace()])
    return text
```

```python
train_data['text'] = train_data['text'].apply(preprocess_text)
test_data['text'] = test_data['text'].apply(preprocess_text)
```

```python

# Feature extraction
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(train_data['text'])
y_train = train_data['class']  # Assuming label column has 'pos' and 'neg'
X_test = vectorizer.transform(test_data['text'])
y_test = test_data['class']
```

```python

# Train the Multinomial Naive Bayes model
model = MultinomialNB()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python

print(f'Accuracy: {accuracy}')
print('Classification Report:')
print(report)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords

# Load the training data
#train_data = pd.read_csv('imdb_train.csv')  # Ensure your CSV has the correct path

# Preprocessing function to remove stop words
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Tokenize the text
    words = text.split()
    return words

# Get the list of stop words
stop_words = set(stopwords.words('english'))

# Tokenize and filter stop words for all reviews in the training set
all_words = []
for review in train_data['text']:
    words = preprocess_text(review)
    filtered_words = [word for word in words if word not in stop_words]
    all_words.extend(filtered_words)

# Create a set of unique words (vocabulary)
vocabulary = set(all_words)

# Size of the vocabulary after removing stop words
vocabulary_size = len(vocabulary)

print(f'The size of the vocabulary after removing stop words is: {vocabulary_size}')
```
