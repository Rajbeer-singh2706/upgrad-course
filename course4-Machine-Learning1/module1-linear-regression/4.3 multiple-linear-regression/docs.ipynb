```python
'''
R-squared vs Adjusted R-squared
Why do you think it is better to use adjusted R-squared in the case of multiple linear regression?
'''

#
'''
In the case of multiple linear regression, **Adjusted R-squared** is often preferred over **R-squared** because it provides a more accurate measure of the model's goodness of fit when multiple predictors are involved. Here's why:

### 1. **R-squared**:
   - **Definition**: R-squared represents the proportion of the variance in the dependent variable that is explained by the independent variables in the model.
   - **Limitation**: R-squared always increases or remains the same when additional predictors are added, even if the new predictors have little to no explanatory power. This can make R-squared misleading in evaluating the true performance of the model.

### 2. **Adjusted R-squared**:
   - **Definition**: Adjusted R-squared adjusts the R-squared value for the number of predictors in the model. It penalizes the addition of unnecessary predictors that don’t significantly improve the model.
   - **Reason for Use in Multiple Linear Regression**: In multiple regression, adding more variables can artificially inflate the R-squared, giving the illusion of a better model. Adjusted R-squared accounts for the number of predictors and helps determine whether the additional predictors genuinely improve the model’s explanatory power.
   - **Formula**:  
     \[
     \text{Adjusted R}^2 = 1 - \left( \frac{(1 - R^2)(n - 1)}{n - p - 1} \right)
     \]
     Where:
     - \(n\) = number of observations
     - \(p\) = number of predictors

   - **Interpretation**: If Adjusted R-squared increases when a new predictor is added, it suggests the new predictor has improved the model. If it decreases, the predictor is likely irrelevant.

In summary, **Adjusted R-squared** is a more reliable metric for evaluating multiple linear regression models because it adjusts for the number of predictors and avoids overfitting by penalizing unnecessary complexity.
'''
## 
'''
Suggested Answer
The major difference between R-squared and Adjusted R-squared is that R-squared doesn't penalise the model for having more number of variables. Thus, if you keep on adding variables to the model, the R-squared will always increase (or remain the same in the case when the value of correlation between that variable and the dependent variable is zero). Thus, R-squared assumes that any variable added to the model will increase the predictive power.

Adjusted R-squared on the other hand, penalises models based on the number of variables present in it. So if you add a variable and the Adjusted R-squared drops, you can be certain that that variable is insignificant to the model and shouldn't be used. So in the case of multiple linear regression, you should always look at the adjusted R-squared value in order to keep redundant variables out from your regression model.


'''
```

```python
#Recursive Feature Elimination
##Read this document and answer the following question. 
#Based on your reading, how do you think RFE measures the importance of the variable?
'''
**Recursive Feature Elimination (RFE)** measures the importance of variables by repeatedly fitting a model and ranking features based on their importance to that model. The process works as follows:

### 1. **Initial Model Fitting**:
   - RFE first fits a model (e.g., a linear model or decision tree) to the entire set of features. The model's coefficients or feature importance scores (depending on the type of model) are used to rank the variables.

### 2. **Ranking Feature Importance**:
   - The importance of each feature is determined by the model's internal mechanism. For example:
     - In linear models, feature importance might be measured by the absolute magnitude of the model coefficients.
     - In decision trees, feature importance is usually based on the reduction in the loss function (e.g., Gini index or entropy for classification, MSE for regression) when a feature is used to split the data.

### 3. **Eliminating Least Important Features**:
   - After ranking the features, the least important features are removed (either one at a time or in batches). This reduction is based on the importance scores from the model.
   
### 4. **Recursive Process**:
   - The process is repeated recursively, where the model is re-trained with the remaining features, and the least important ones are pruned in each iteration until the desired number of features is reached.

### How Does RFE Measure Importance?
RFE relies on the feature importance scores derived from the underlying model:
- In linear models, importance is often measured by the magnitude of the feature coefficients.
- In tree-based models, importance might be derived from the information gain or impurity reduction (e.g., how much a feature contributes to improving the model's decision-making process).

The key idea is that RFE eliminates features that contribute the least to the model's predictive power and keeps those that provide the most useful information for making accurate predictions. It does so recursively, ensuring that the most important variables are retained as features are removed.
'''

###
'''
Recursive feature elimination is based on the idea of repeatedly constructing a model (for example, an SVM or a regression model) and choosing either the best or worst performing feature (for example, based on coefficients), setting the feature aside and then repeating the process with the rest of the features. This process is applied until all the features in the dataset are exhausted. Features are then ranked according to when they were eliminated. As such, it is a greedy optimisation for finding the best performing subset of features.


'''
```
